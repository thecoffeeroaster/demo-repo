```python
##importing modules

import plotly.graph_objects as go
import plotly.offline as py
import numpy as np

##Meshgrid
L=1
x = np.linspace(-L, L, 5)
y = np.linspace(-L, L, 5)

print('x: ', x, 'y: ', y)

## (X,Y) is a 'tuple' with components X and Y, each being a matrix of x and y coordinates
(X,Y) = np.meshgrid(x,y)  
print("here is X:")
print(X)
print("here is Y:")
print(Y)
```

```python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

```python
L=5; n=101
x = np.linspace(-L, L, n)
y = np.linspace(-L, L, n)
(X,Y) = np.meshgrid(x,y)

## play around with these functions. Plot them all, one at a time
Z = 3*Y/(X**2 + Y**2 + 1)+2
#Z = -0.2*X + 0.3*Y + 1
#Z = 0.5*np.sin(X**2 + Y**2)/(X**2 + Y**2)
#Z = X*Y/(X**2 + Y**2)

#Z = 1/2 * X**2 - 1/2 * Y**2



## play around with various values of contours, colorscale and showscale (see also Plotly's documentation)
surf = go.Surface(x=X, y=Y, z=Z, 
                 contours = {
                    "x": {"show": True, "start": -L, "end": L, "size": 0.25,
                          "highlightcolor": "darkorange"},
                    "y": {"show": True, "start": -L, "end": L, "size": 0.25,
                          "highlightcolor": "darkorange"},
                    "z": {"show": False, "start": -L, "end": L, "size": 0.1,
                         "usecolormap": True, "highlightcolor": "red"},
                 },
                 colorscale = "Viridis",
                 showscale = False,   #don't show palette of colors on the side 
                 name = r"${3y \over x^2+y^2+1}+2$"
       )

fig = go.Figure(surf)

fig.update_layout(
            title = " My First Plotly Plot",
            scene = {
             "xaxis": {"range": [-L, L]},
             "yaxis": {"range": [-L, L]},
             "zaxis": {"range": [0, L]},
            "camera_eye": {"x": -1, "y": -1.6, "z": 0.3},
            "aspectratio": {"x": 1, "y": 1, "z": 1} 
            }
)

fig.update_layout(
    hovermode=False,   #don't show coordinates when hovering
    autosize=False,
    width=700,
    height=600,
    margin=dict(
        l=10,
        r=10,
        b=20,
        t=50
    )
)


fig.show()
configure_plotly_browser_state()

```
